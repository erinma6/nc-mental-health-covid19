```python
from cx100 import refined
%matplotlib inline
from matplotlib import pyplot
import numpy
```

```python
# Percentage of excessive alcohol use in NC by year

alcohol = [["Year", "Percentage of Excessive Alcohol Population"], [2015, 13], [2016, 15], [2017, 15], [2018, 17], [2019, 17], [2020, 18], [2021, 18]]
refined(alcohol)
```

```python
# Percentage of excessive alcohol use in NC by year

years = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
alcohol = [13, 15, 15, 17, 17, 18, 18]
pyplot.plot(years, alcohol)
pyplot.xlabel("Years")
pyplot.ylabel("Percentage of Population")
pyplot.title("Excessive Alcohol Use in NC by Year");
```

```python
# Number of bad mental health days in NC by year

years = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
mental_health_days = [3.4, 3.7, 3.7, 3.9, 3.9, 4.1, 4.3]
pyplot.plot(years, mental_health_days);
```

```python
# Ratio of mental health providers to the population


years = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
mental_health_providers = [1/472, 1/440, 1/490, 1/460, 1/440, 1/410, 1/390]
pyplot.plot(years, mental_health_providers);
```

```python
# Particulate matter in microorganisms per cubic meter (level of air pollution)

years = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
air_pollution = [12.3, 12.3, 9.1, 9.1 , 9.8, 9.5, 8.5]
pyplot.plot(years, air_pollution);
```

```python
# Emergency Department Visit Rate for Drug Overdose

years = [2016, 2017, 2018, 2019, 2020, 2021]
air_pollution = [105.7, 127, 115.6, 115.9, 142.5, 150.8]
pyplot.plot(years, air_pollution)
pyplot.xlabel("Years")
pyplot.ylabel("Visit Rate")
pyplot.title("Emergency Department Visit Rate for Drug Overdose [per 100,000 residents]");
```
